```python
import pandas as pd
import numpy as np
dataframe = pd.read_csv("results.csv", header=None)
```

```python
population_size = 100
num_generations = int((len(dataframe.index)/population_size))
mean_values = []
max_values = []
min_values = []
for i in range(1,num_generations):
    generation = dataframe[(i-1)*population_size:i*population_size]
    mean_values.append(generation[generation.columns[10]].mean())
    max_values.append(generation[generation.columns[10]].max())
    min_values.append(generation[generation.columns[10]].min())
x = range(1,num_generations)
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()
ax.plot(x, mean_values)
ax.plot(x, max_values)
ax.plot(x, min_values)
```
